```python
import pandas as pd
import numpy as np
```

```python
store = pd.read_csv('retail_store_sales.csv')
store.isna().sum()
```

```python

unit_mask = store['Total Spent'].notna() & store['Quantity'].notna()

store.loc[unit_mask, 'Price Per Unit'] = store.loc[unit_mask, 'Total Spent'] / store.loc[unit_mask, 'Quantity']

store = store.dropna(subset=['Quantity', 'Total Spent'])

store['Item'] = store['Item'].fillna(store['Category'])

store['Discount Applied'] = store['Quantity'] * store['Price Per Unit'] > store['Total Spent']

store.isna().sum()
```

```python
store.to_csv('cleaned_retail.csv', index=False)

```
